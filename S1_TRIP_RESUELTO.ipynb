# Paso 1: Evaluación de Calidad de Datos

Store 1 tiene como objetivo garantizar la coherencia en la recopilación de datos. Como parte de esta iniciativa, se debe evaluar la calidad de los datos recopilados sobre los usuarios y las usuarias. Te han pedido que revises los datos recopilados y propongas cambios. A continuación verás datos sobre un usuario o una usuaria en particular; revisa los datos e identifica cualquier posible problema.

```python
user_id = '32415'
user_name = ' mike_reed '
user_age = 32.0
fav_categories = ['ELECTRONICS', 'SPORT', 'BOOKS']
```

**Opciones:**

1. Se debe cambiar el tipo de dato de `user_id` de cadena a entero.
    
2. La variable `user_name` contiene una cadena que tiene espacios innecesarios y un guion bajo entre el nombre y el apellido.
    
3. El tipo de dato de `user_age` es correcto y no hay necesidad de convertirlo.
    
4. La lista `fav_categories` contiene cadenas en mayúsculas. En lugar de ello, debemos convertir los valores de la lista en minúsculas.

**Respuesta del Analista:**

**Opción 1 - NO es un problema real:**
El `user_id` debe mantenerse como cadena (string). Aunque contiene solo dígitos, los IDs no son valores numéricos sobre los que vayamos a realizar operaciones matemáticas. Mantenerlo como string nos permite:
- Preservar ceros a la izquierda si los hubiera
- Evitar problemas de precisión numérica
- Facilitar búsquedas y concatenaciones

**Opción 2 - SÍ es un problema real:**
El nombre contiene espacios innecesarios al inicio y al final (` mike_reed `), además del guion bajo que debe ser reemplazado por un espacio. Esto dificulta:
- Búsquedas y comparaciones exactas
- Presentación de datos al usuario
- Análisis de texto y normalización

**Opción 3 - NO es correcto:**
El tipo de dato `float` (32.0) NO es apropiado para la edad. Las edades son números enteros, y usar float:
- Desperdicia memoria
- Puede crear confusión (¿32.5 años?)
- No tiene sentido semántico
Debe convertirse a `int`.

**Opción 4 - NO es necesario:**
Las categorías en mayúsculas son una convención válida para constantes o identificadores estandarizados. No hay necesidad de convertirlas a minúsculas a menos que:
- Exista un estándar de negocio específico
- Haya inconsistencias en la capitalización entre registros
En este caso, mantener mayúsculas facilita la distinción de categorías como datos estructurados.

# Paso 2: Corrección de user_name

Vamos a implementar los cambios que identificamos. Primero, necesitamos corregir los problemas de la variable `user_name`. Como vimos, tiene espacios innecesarios y un guion bajo como separador entre el nombre y el apellido; tu objetivo es eliminar los espacios y luego reemplazar el guion bajo con el espacio.

```python
user_name = ' mike_reed '
user_name = user_name.strip()  # Eliminar espacios al inicio y al final
user_name = user_name.replace('_', ' ')  # Reemplazar guion bajo con espacio

print(user_name)
```

```python
user_name = 'mike reed'
name_split = user_name.split()  # Dividir por espacios

print(name_split)
```

# Paso 10: Limpieza Completa de Datos

Apliquemos ahora todos los cambios a la lista de clientes. Para simplificar las cosas, te proporcionaremos una más corta.

Debes:
1. Eliminar todos los espacios iniciales y finales de los nombres, así como cualquier guion bajo.
2. Convertir todas las edades en números enteros.
3. Separar todos los nombres y apellidos en una sublista.

Guarda la lista modificada como una nueva lista llamada `users_clean` y muéstrala en la pantalla.

```python
users = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['CLOTHES', 'BOOKS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRONICS', 'HOME', 'FOOD'], [459, 120, 99]],
]

users_clean = []

# Procesa al primer usuario
user_name_1 = users[0][1].strip().replace('_', ' ')  # Limpiar nombre
user_age_1 = int(users[0][2])  # Convertir edad a entero
user_name_1 = user_name_1.split()  # Dividir nombre en lista
users_clean.append(
    [users[0][0], user_name_1, user_age_1, users[0][3], users[0][4]])

# Procesa al segundo usuario
user_name_2 = users[1][1].strip().replace('_', ' ')
user_age_2 = int(users[1][2])
user_name_2 = user_name_2.split()
users_clean.append(
    [users[1][0], user_name_2, user_age_2, users[1][3], users[1][4]])

# Procesa al tercer usuario
user_name_3 = users[2][1].strip().replace('_', ' ')
user_age_3 = int(users[2][2])
user_name_3 = user_name_3.split()
users_clean.append(
    [users[2][0], user_name_3, user_age_3, users[2][3], users[2][4]])

print("\nDatos limpios:")
for user in users_clean:
    print(
        f"ID: {user[0]}, Nombre: {user[1]}, Edad: {user[2]}, Categorías: {user[3]}, Gastos: {user[4]}")
```

## Bonus: Solución Optimizada con Bucle

Como analista de datos, propongo una solución más escalable usando un bucle:

```python
users = [
    ['32415', ' mike_reed ', 32.0, ['ELECTRONICS', 'SPORT', 'BOOKS'], [894, 213, 173]],
    ['31980', 'kate morgan', 24.0, ['CLOTHES', 'BOOKS'], [439, 390]],
    ['32156', ' john doe ', 37.0, ['ELECTRONICS', 'HOME', 'FOOD'], [459, 120, 99]],
]

users_clean = []

for user in users:
    user_id = user[0]
    user_name = user[1].strip().replace('_', ' ').split()
    user_age = int(user[2])
    fav_categories = user[3]
    spendings = user[4]

    users_clean.append(
        [user_id, user_name, user_age, fav_categories, spendings])

print("\n" + "="*70)
print("RESUMEN DE DATOS LIMPIOS")
print("="*70)

for i, user in enumerate(users_clean, 1):
    print(f"\nUsuario {i}:")
    print(f"  ID: {user[0]}")
    print(f"  Nombre completo: {' '.join(user[1]).title()}")
    print(f"  Edad: {user[2]} años")
    print(f"  Categorías favoritas: {', '.join(user[3])}")
    print(f"  Gastos por categoría: {user[4]}")
    print(f"  Gasto total: ${sum(user[4])}")
```
